```python
import tkinter as tk
import pandas as pd
import numpy as np
from tkinter import *
```

```python
root=tk.Tk()
root.config(bg="white")
root.geometry('800x700')
root.title("PredictHeartDisease")

model = pd.read_pickle('PredictHeartDisease')

tk.Label(root,text='Predict Heart Disease',bg='crimson',fg='White',font=('goudy old style',15,'bold')).place(x=0,y=0,relwidth=1)
tk.Label(root,width=20,text="Age",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=50)
tk.Label(root,width=20,text="Sex",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=90)
tk.Label(root,width=20,text="Chest Pain Type",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=130)
tk.Label(root,width=20,text="Rest BP",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=170)
tk.Label(root,width=20,text="Cholestoral",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=210)
tk.Label(root,width=20,text="F blood sugar",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=250)
tk.Label(root,width=20,text="Rest ECG",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=290)
tk.Label(root,width=20,text="Enaga",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=330)
tk.Label(root,width=20,text="Oldpeak",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=370)
tk.Label(root,width=20,text="Slope",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=410)
tk.Label(root,width=20,text="No.of Major Vessels",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=450)
tk.Label(root,width=20,text="Thal",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=490)
tk.Label(root,width=20,text="max_heart_rate",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=0,y=530)


age = tk.Variable(root)
sex = IntVar(root)
sex.set("Choose Sex")

chest_py = IntVar(root)
chest_py.set("Chest Pain")

rest_bp = tk.Variable(root)
cholestrol = tk.Variable(root)
fast_bs = IntVar(root)
fast_bs.set("Blood Sugar")
rest_ecg = IntVar(root)
rest_ecg.set("Rest ECG")
max_heart_rate = tk.Variable(root)
exang = tk.Variable(root)
oldpeak = tk.Variable(root)
slope = IntVar(root)
slope.set("Slope")
major_vessels = IntVar(root)
major_vessels.set("Vessels flourosopy")
thal = IntVar(root)
thal.set("thal")
predict_var = tk.Variable(root)

tk.Entry(root,width=25,textvariable=age,font=('Arial',10),bg='light yellow').place(x=250,y=50)
w = OptionMenu(root, sex, "1", "0").place(x=250,y=90)
w = OptionMenu(root, chest_py, "1", "2", "3", "4").place(x=250,y=130)
tk.Entry(root,width=25,textvariable=rest_bp,font=('Arial',10),bg='light yellow').place(x=250,y=170)
tk.Entry(root,width=25,textvariable=cholestrol,font=('Arial',10),bg='light yellow').place(x=250,y=210)
w = OptionMenu(root, fast_bs, "0","1").place(x=250,y=250)
w = OptionMenu(root, rest_ecg, "0","1").place(x=250,y=290)
tk.Entry(root,width=25,textvariable=exang,font=('Arial',10),bg='light yellow').place(x=250,y=330)
tk.Entry(root,width=25,textvariable=oldpeak,font=('Arial',10),bg='light yellow').place(x=250,y=370)
w = OptionMenu(root, slope, "0","1","2").place(x=250,y=410)
w = OptionMenu(root, major_vessels, "0","1","2","3","4").place(x=250,y=450)
w = OptionMenu(root, thal, "3","6","7").place(x=250,y=490)
tk.Entry(root,width=25,textvariable=max_heart_rate,font=('Arial',10),bg='light yellow').place(x=250,y=530)


tk.Label(root,textvariable=predict_var,fg="black",bg="#F0F8FF",font=("Helvetica", 15,"bold")).place(x=180,y=650)
def predict():
    query = pd.DataFrame({
        'age' : [age.get()],
        'sex' : [sex.get()],
        'cp' : [chest_py.get()],
        'trestbps' : [rest_bp.get()],
        'chol' : [cholestrol.get()],
        'fbs' : [fast_bs.get()],
        'restecg' : [rest_ecg.get()],
        'thalach' : [max_heart_rate.get()],
       'exang' : [exang.get()],
        'oldpeak' : [oldpeak.get()],
        'slope' : [slope.get()],
        'ca' : [major_vessels.get()],
        'thal' : [thal.get()],
    })
    
    if (model.predict(query)[0]==1):
        predict_var.set("Patient is suffering with heart disease, will soon recover!")
    else:
        predict_var.set("Patient don't have heart disease, stay healthy!")

    
# w.pack()
tk.Button(root,text="Predict",fg="black",bg="#F0F8FF",font='arial 15 underline',relief="groove",command=predict).place(x=300,y=580) 

root.mainloop()
```
